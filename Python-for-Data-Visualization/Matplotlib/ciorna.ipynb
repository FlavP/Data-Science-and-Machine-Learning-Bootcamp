```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.plot(x,z)

ax2 = fig.add_axes([0.2, 0.6, 0.2, 0.3])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
ax2.set_title('Zoom')
ax2.plot(x,y)
```
